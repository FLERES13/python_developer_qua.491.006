A Cyberdyne Systems contratou cientistas de dados do SENAI *experts* na liguagem de programação python para analisar qual indice de cancelamentos de contratos por parte dos clientes da empresa. O desafio aqui é tentar descobrir a causa do alto Indice de cancelamos e indidar a melhor solução para a empresa.

Para isso , a Cyberdyne Systems forneceu um *dataset* com os detalhes dos cancelamentos, Os cientistas devem analisar o *dataset* através de biblioteca ***Pandas***

Vamos começar trazendo todos os dados do *dataset* e fazendo o tratamento de dados 

```python
import pandas as pd

tabela = pd.read_csv("cancelamentos.csv")

display(tabela)
```

Agora vou eliminar a coluna de id e mandar re-exibir a tabela. 

```python
tabela = tabela.drop("CustomerID", axis=1)
display(tabela)
```

Vamos pedir para que o Phyton nos retorne o número de linhas preenchidas da tabela 

```python
display(tabela.info())
```

Veja que temos em cada uma das colunas um número diferente de linhas. Isso acontece porque temos dados nulos da tabela, e que devemos eliminar.

```python
tabela = tabela.dropna()
display(tabela.info())
```

Agora, a análise de dados propriamente dita vai começar, Vamos verificar o número de clientes que cancelaram o contrato e o número de clientes que não cancelaram

```python
display(tabela["cancelou"].value_counts())
```

Sabemos que o número de cancelamos é maior do que o número que não cancelaram, mas quanto isso dá em porcentagem ? Vamos descobrir 

```python
display(tabela["cancelou"].value_counts(normalize=True).map("{:.2%}".format))
```

Sabemos agora que aproximadamente 56% dos clientes cancelaram, o que é um número muito alto.Vamos descobrir o por que.


```python
display(tabela["duracao_contrato"].value_counts())
display(tabela["duracao_contrato"].value_counts(normalize=True).map("{:.2%}".format))
```

Podemos perceber que existe uma discrepância entre contratos do tipo mensal e dos outros tipos, Vamos analisar a taxa de cancelamento de cada tipo de duração de contrato

```python
display(tabela.groupby("duracao_contrato").mean(numeric_only=True))
```

Logo de cara , podemos ver que todos os clientes que optaram pelo tipo de contrato mensal cancelaram o serviço, Em outras palavras, contrato mensal teve 100% de cancelamos.

## O que pode ser feito ?

Podemos retirar o tipo de contrato mensal de análise para verificar qual porcentagem de cancelamemto sobra. 

```python
tabela = tabela[tabela["duracao_contrato"] != "Monthy"]
display(tabela)
display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize=True).map("{:.2%}".format))
```

Pudemos verificar que sem o contrato de tipo mensal a taxa de cancelamemtos diminuiu para 46,05%, uma queda de mais de 10% na taxa de cancelamentos. O problema é que a taxa de cancelamentos continua muito alta, afinal , nenhuma empresa quer uma taxa de cancelamentos perto da metade, então precisamos continuar com a analise para verificar outras possíveis razãoes para a alta taxa de cancelamentos. 

Vamos analisar graficamento falando cada uma das colunas. Ou seja, Vamos criar graficos de cada coluna para verificar de ha discrepancias muito grandes nas taxas de cancelamentos em alguma dessas colunas

```python
import plotly.express as px

for coluna in tabela.columns:
    if coluna != "cancelou":  # evita fazer 'cancelou' vs 'cancelou'
        grafico = px.histogram(tabela, x=coluna, color="cancelou", width=600)
        grafico.show()
```

Os graficos indicam problemas com os usuários que fazem mais de 5 ligações para o *call center*, e também cancelametos por parte dos usuários que passam de 20 dias inadimplentes.Portanto, deveremos eliminar esses clientes de análise a fim de verificar a nova taxa de cancelamento sem esses clientes.

```python
tabala = tabela[tabela["ligacoes_callcenter"] < 5]
tabela = tabela[tabela["dias_atraso"] <=20]
display(tabela)
display(tabela["cancelou"].value_counts())
display(tabela["cancelou"].value_counts(normalize=True).map("{:.2%}".format))
```

Finalmente conseguimos diminuir a taxa de cancelamentos para 18,4%.
